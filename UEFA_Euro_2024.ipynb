## Audience Analysis:
Audience Attendance: Calculate the average audience attendance per match or stadium.

```python
# Convert the 'audience' column to numeric type by removing commas and casting to integer

if matches['audience'].dtype == object:
    matches['audience'] = matches['audience'].str.replace(',', '').astype(int)
avg_audience_per_match = matches['audience'].mean()
print(avg_audience_per_match)

```

```python
avg_audience_per_stadium =matches.groupby('stadium')['audience'].mean()
avg_audience_per_stadium
```

# **Shot Analysis**

Shots and Accuracy: Analyze the shooting accuracy by comparing total shots to shots on target for both teams

```python
matches.head(2)
```

```python
matches['team1_shot_accuracy'] = matches['team1_shots_on_target'] / matches['team1_total_shots']
matches['team2_shot_accuracy'] = matches['team2_shots_on_target'] / matches['team2_total_shots']
matches.head(2)
```
