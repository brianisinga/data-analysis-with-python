```python
##treating missing values
data['Urea'].fillna(data['Urea'].mean(), inplace=True)
data['AGE'].fillna(data['AGE'].mode(), inplace=True)
data['Cr'].fillna(data['Cr'].median(), inplace=True)
data['HbA1c'].fillna(data['HbA1c'].mean(), inplace=True)
data['Chol'].fillna(data['Chol'].mean(), inplace=True)
data['TG'].fillna(data['TG'].mean(), inplace=True)
data['HDL'].fillna(data['HDL'].mean(), inplace=True)
data['LDL'].fillna(data['LDL'].mean(), inplace=True)
data['VLDL'].fillna(data['VLDL'].mean(), inplace=True)
```

```python
# List of columns to plot
columns = ['Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL']

# Number of subplots
num_plots = len(columns)

# Create subplots
fig, axes = plt.subplots(nrows=num_plots, ncols=1, figsize=(10, 20))

# Plot each column
for i, col in enumerate(columns):
    axes[i].boxplot(female[col])
    axes[i].set_title(f'{col} Boxplot')
    axes[i].set_ylabel(col)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Define a function to detect outliers
def detect_outliers(data, column, threshold=2.5):
    z_scores = np.abs((data[column] - data[column].mean()) / data[column].std())
    return z_scores > threshold

# Detect outliers for 'Urea' and 'Cr'
outliers_urea = detect_outliers(female, 'Urea')
outliers_cr = detect_outliers(female, 'Cr')

# Print outliers
print("Outliers for urea:\n", outliers_urea)
print("\nOutliers for Cr:\n", outliers_cr)
```
